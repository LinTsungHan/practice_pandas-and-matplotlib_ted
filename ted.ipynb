```python
#處理時間 
from datetime import datetime
#datetime裡面的utcfromtimestamp套件,將電腦時間轉英國格林威治時間
str(datetime.utcfromtimestamp(1140825600))
```

```python
#轉換流程 -> 遞給整行 ->每一每一個走過流程
#def timetrans(input):定義轉化流程
def timetrans(s):
    return str(datetime.utcfromtimestamp(s))
#apply :apply(流程名字) 對每一個執行一次流程
df["filmutc"] = df["film_date"].apply(timetrans)
df
```

```python
#篩選操作:原理 df[ [跟你的列一樣多的Boolean] ]
#靠轉換操作去轉換
def fflow(n):
    if n>3000 and n<5000:
        return True
    else:
        return False
f = df["comments"].apply(fflow)
df[f]    
```

```python
#eval是一個把傳入的str當成一個py程式在執行
#ex
eval("print(3 +2)")
eval("[1,2,3]")
```

```python
#轉換字串成list
def fflow(s):
    taglist = eval(s)
    if "children" in taglist:
        return True
    else:
        return False
    
df[ df["tags"] .apply(fflow)]
```
